```python
import pandas as pd
import os
```

```python
path = os.getcwd()
path
```

```python
predict = pd.read_csv("gpt_predict_v2.csv", encoding="utf-8-sig")
predict
```

```python
df = pd.read_excel("APEACH_데이터_라벨링완료v1.xlsx")
df
```

```python
df['is_immoral'] = ~df['is_immoral']
df
```

```python
df['label_numbers'] = df['label_numbers'].str.replace(r'\[|\]', '', regex=True)
df
```

```python
def process_labels(row):
    if row['is_immoral']:
        # True인 경우 기존 로직 적용
        return row['label_numbers']
    else:
        # False인 경우 [9]로 설정
        return 9
```

```python
df['label_numbers'] = df.apply(process_labels, axis=1)
df
```

```python
# df['label_numbers'] = df['label_numbers'].apply(lambda x: int(x.strip("[]")) if isinstance(x, str) and x.strip("[]").isdigit() else None)
```

```python
df['predicted'] = predict['label']
df
```

```python
data = {
    'label': df['label_numbers'],
    'predicted': predict['label']
}
df_test = pd.DataFrame(data)
df_test
```

```python
df_test = df_test.dropna()
```

```python
df_test['label'] = df_test['label'].apply(lambda x: int(x.split(',')[0]) if isinstance(x, str) else x)
df_test['predicted'] = df_test['predicted'].apply(lambda x: int(x.split(',')[0]) if isinstance(x, str) else x)

# 정수 비교
df_test['correct'] = df_test['label'] == df_test['predicted']
df_test
```

```python
accuracy = df_test['correct'].mean()
print(f"정확도: {accuracy * 100:.2f}%")
```

```python
correct_count = df_test['correct'].sum()
incorrect_count = len(df) - correct_count
print(f"정확히 분류된 샘플 수: {correct_count}")
print(f"잘못 분류된 샘플 수: {incorrect_count}")
```

```python
class_accuracy = df_test.groupby('label')['correct'].mean()
class_accuracy
```
