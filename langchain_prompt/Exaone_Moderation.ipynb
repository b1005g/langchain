```python
# 베이스 모델 로드
base_model_id = "LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct"
base_model = AutoModelForCausalLM.from_pretrained(base_model_id, trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained(base_model_id)

# 어댑터 모델 로드
adapter_model_path = "./EXAONE-3.5-2.4B-Instruct_finetuning_v2"  # adapter_config.json과 adapter_model.safetensors 위치
model = PeftModel.from_pretrained(base_model, adapter_model_path)

# 필요한 토큰 설정
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"
```

```python
train = pd.read_csv("Real_Test.csv", encoding ='utf-8-sig')
train
```

```python
train_text = train['text']
train_text
```

```python
pipe = pipeline(
    task="text-generation",
    model=model, 
    tokenizer=tokenizer 
)
change_texts = []
categories = [
    "폭언 아님(0)", "폭언(1)", "단순욕설 (2)", "외모비하 (3)", "장애인 차별 (4)", "인종차별 (5)", "종교차별 (6)", 
    "지역차별 (7)", "성차별 (8)", "나이 (9)", "협박 (10)", "성희롱 (11)"
]
for text in train_text:
    query = text
    messages = [
        {
            "role": "system",
            "content": (
                f"다음 문장을 읽고, 해당 문장이 아래 카테고리 중 어떤 것에 해당하는지 판별하세요:\n"
                f"카테고리: {', '.join(categories)}\n\n"
                f"답변 형식: 카테고리 번호를 얘기하고, 0이 아니라면 텍스트를 바꿔주세요"
                f"예: '10 협박, 순화된 문장: 너희를 무서워하게 하려는 것이 아니야.'\n"  
            )
        },
        {
            "role": "user",
            "content": f"input : {query}, output : "
        },
    ]
    prompt = "\n".join([m["content"] for m in messages]).strip()
    outputs = pipe(
        prompt,
        do_sample=True,
        temperature=0.2,
        top_p=0.9,
        max_new_tokens=len(query) + 10,
    )
    
    generated_text = outputs[0]['generated_text']
    result = generated_text.split("output :")[-1].strip() 
    #result = generated_text[len(prompt):].strip()
    # result = generated_text.split("output :")[-1].strip()
    print(result)
    change_texts.append(result)
```
