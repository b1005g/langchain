```python
import pandas as pd
import os
import json
from sklearn.model_selection import train_test_split
```

```python
#train.to_csv("cohere_train.csv",encoding="utf-8-sig",index=False)
```

```python
train = pd.read_csv("데이터_최종본_진짜임.csv", encoding ='utf-8-sig')
train
```

```python
train['label_words'].value_counts()
```

```python
# 1) 각 라벨의 등장 횟수 구하기
label_counts = train['label_words'].value_counts()

# 2) 라벨이 5개 이상 등장하는 것만 필터링
valid_labels = label_counts[label_counts >= 5].index

# 3) 해당 라벨만 남기기
train_filtered = train[train['label_words'].isin(valid_labels)]
```

```python
train_filtered
```

```python
train_filtered = train_filtered.rename(columns={'label_numbers' : 'label'})
train_filtered
```

```python
df_train, df_test = train_test_split(train_filtered, test_size=0.2, random_state=42)
```

```python
df_train
```

## 데이터 만들고 보내기

```python
for row in df_train.itertuples():
    print(row)
```

```python
def create_classification_data(text, labels):
    if isinstance(labels, str):
        labels_list = [lbl.strip() for lbl in labels.replace('[','').replace(']','').split(',')]
    else:
        labels_list = labels
    formatted_data = {
        "text": text,
        "label": labels_list
    }
    return formatted_data

if not os.path.isfile("data.jsonl"):
    print("Creating jsonl file ...")
    with open("cohere_data_v3.jsonl", 'w+') as file:
        for row in df_train.itertuples():
            formatted_data = create_classification_data(row.text, row.label)
            file.write(json.dumps(formatted_data) + '\n')
        file.close()
        print("Done")
else:
    print("coheredata_v3.jsonl file already exists")
```

```python
# df_test jsonl 만들기
if not os.path.isfile("data.jsonl"):
    print("Creating jsonl file ...")
    with open("cohere_data_v3_test.jsonl", 'w+') as file:
        for row in df_train.itertuples():
            formatted_data = create_classification_data(row.text, row.label)
            file.write(json.dumps(formatted_data) + '\n')
        file.close()
        print("Done")
else:
    print("coheredata_v3_test.jsonl file already exists")
```

```python
import cohere
#from cohere.classify import Example

co = cohere.Client('bQJAFwtllUOZ91VnHYttBLAkxrTGVHXyNKSPgcCc') # This is your trial API key
response = co.classify(
  model='aa14a851-2c24-4754-ad1b-f545262cdb2c-ft',
  inputs=["엄마가 죽었나봐"])

for classification in response.classifications:
    print(f"Text: {classification.input}")
    print(f"Prediction: {classification.prediction}")
    print(f"Confidence: {classification.confidence}")
```

```python
def classify_text_finetune(texts):
    classifications = co.classify(
        model='aa14a851-2c24-4754-ad1b-f545262cdb2c-ft',
        inputs=texts,
        examples=None
    ).classifications
    return [c.predictions for c in classifications]

# Create batches of texts and classify them
BATCH_SIZE = 90 # The API accepts a maximum of 96 inputs
y_pred = []
for i in range(0, len(df_test), BATCH_SIZE):
    batch_texts = df_test["text"][i:i+BATCH_SIZE].tolist()
    y_pred.extend(classify_text_finetune(batch_texts))
```

```python
y_pred
```

```python
y_pred = [labels if labels else ['0'] for labels in y_pred]
y_pred
```

```python
df_test_label = df_test["label"].tolist()
df_test_label
```

```python
import re
parsed_data = []
for s in df_test_label:
    # 대괄호 안의 숫자만 추출 => ['9','3'] 등
    nums = re.findall(r'\d+', s)  
    parsed_data.append(nums)

print(parsed_data)
```

```python
parsed_data
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
y_true_list = parsed_data
y_pred_list = y_pred
```

```python
y_true_bin = mlb.fit_transform(y_true_list)
y_pred_bin = mlb.transform(y_pred_list)
```

```python
accuracy = accuracy_score(y_true_bin, y_pred_bin)
f1 = f1_score(y_true_bin, y_pred_bin, average='weighted')

print(f'Accuracy: {100*accuracy:.2f}')
print(f'F1-score: {100*f1:.2f}')
```

# 내가 직접 파인튜닝하기

```python
import cohere
co = cohere.Client('bQJAFwtllUOZ91VnHYttBLAkxrTGVHXyNKSPgcCc')
response = co.finetuning.list_finetuned_models()
print(response)
```

```python
from cohere.finetuning import (
    BaseModel,
    FinetunedModel,
    Hyperparameters,
    Settings,
)
hp = Hyperparameters(
    early_stopping_patience=10,
    early_stopping_threshold=0.001,
    train_batch_size=16,
    train_epochs=3,
    learning_rate=0.01,
)
# Fine-tuning 시작
finetuned_model = co.finetuning.create_finetuned_model(
    request=FinetunedModel(
        name="test-finetuned-model_0126_v1",
        settings=Settings(
            base_model=BaseModel(
                base_type="BASE_TYPE_CLASSIFICATION",
            ),
            dataset_id="multi-label-dataset-e5frzt",
            hyperparameters=hp,
        ),
    )
)
print(finetuned_model)
```
